### Show the most number of tv-shows and movies released by each country in descending order in Netflix dataset shared in class. Also plot a count plot of each country.

```python
data = pd.read_csv('netflixData.csv')
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
sb.heatmap(data.isnull())
```

```python
data['Title'][2:4]
```

```python
data[(data['Production Country'] == 'United States')].head()
```

```python
shows_movies_data = data['Production Country'].value_counts().sort_values(ascending=False)
shows_movies_data.head()
```

```python
show_movies_df = pd.DataFrame(data=shows_movies_data)
show_movies_df
```

```python
sb.countplot(shows_movies_data)
```
